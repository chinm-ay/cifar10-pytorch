```python
import json
import os

import pandas as pd
import matplotlib.pyplot as plt
```

```python
experiments = [
    'convmixer256_8_k5_p1_00', 'convmixer256_8_k5_p1_01', 'convmixer256_8_k5_p1_02',
    'convmixer256_8_k5_p2_00', 'convmixer256_8_k5_p2_01', 'convmixer256_8_k5_p2_02',
    # 'convmixer256_8_k9_p1_00', 'convmixer256_8_k9_p1_01','convmixer256_8_k9_p1_02',
    'convmixer256_8_k9_p2_00', 'convmixer256_8_k9_p2_01', 'convmixer256_8_k9_p2_02',
                                                          'convmixer256_16_k9_p2_00'
]
final_results = 'convmixer256_8_k5_p2_05'
resnset_s_results = 'resnet_s38_default'
```

```python
data = []
for _, experiment in enumerate(experiments):
    print(experiment)
    df = pd.read_json(f"../logs/{experiment}/train_{experiment}").T
    data.append(df)
```

```python
plt.figure()
sample = experiments[3:6]
labels = ["Aug lvl 0", "Aug lvl 1", "Aug lvl 2"]
for idx, experiment in enumerate(sample):
    plt.plot(data[3:6][idx].train_acc, label=labels[idx])
    print(f"{labels[idx]} max: {data[3:6][idx].train_acc.max()}")
plt.legend()
plt.xlabel('Epoch')
plt.xlim(0,99)
plt.ylabel('Training accuracy')
plt.ylim(top=1.0)
plt.suptitle("Training accuracy over 100 epochs")
plt.title("(h=256, d=8, k=5, p=2)")
plt.savefig("../publish/images/training_k5pt2.png", dpi=300)
plt.show()
```

```python
plt.figure()
sample = experiments[::3]
labels = ['d=8, k=5, p=1', 'd=8, k=5, p=2', 'd=8, k=9, p=1', 'd=16, k=9, p=2']
for idx, experiment in enumerate(sample):
    plt.plot(data[::3][idx].val_acc, label=labels[idx])
    print(f"{labels[idx]} max: {data[::3][idx].val_acc.max()}")
plt.legend()
plt.xlabel('Epoch')
plt.xlim(0,99)
plt.ylabel('Accuracy')
plt.ylim(top=1.0)
plt.suptitle("Validation accuracy over 100 epochs")
plt.title("(h=256, Lvl 0 Augmentation)")
plt.savefig("../publish/images/val_lvl0.png", dpi=300)
plt.show()
```

```python
plt.figure()
sample = experiments[2::3]
labels = ['k=5, p=1', 'k=5, p=2', 'k=9, p=1']
for idx, experiment in enumerate(sample):
    plt.plot(data[2::3][idx].val_acc, label=labels[idx])
    print(f"{labels[idx]} max: {data[2::3][idx].val_acc.max()}")
plt.legend()
plt.xlabel('Epoch')
plt.xlim(0,99)
plt.ylabel('Accuracy')
plt.ylim(top=1.0)
plt.suptitle("Validation accuracy over 100 epochs")
plt.title("(h=256, d=8, Lvl 2 Augmentation)")
plt.savefig("../publish/images/val_lvl2.png", dpi=300)
plt.show()
```

```python
data_final = []

df = pd.read_json(f"../logs/{final_results}/train_{final_results}").T
data_final.append(df)
df = pd.read_json(f"../logs/{resnset_s_results}/train_{resnset_s_results}").T
data_final.append(df)
```

```python
data_final[0]
```

```python
plt.figure()
samples = [final_results, resnset_s_results]
labels =['ConvMixer256/8 (k=5, p=2)', 'ResNet/S-38 (n=6)']
for idx, experiment in enumerate(samples):
    plt.plot(data_final[idx].train_acc, label=f"{labels[idx]} accuracy")
    print(f"{labels[idx]} max: {data_final[idx].train_acc.max()}")
    loss = data_final[idx].train_loss
    loss_normalized = (loss - loss.min()) / (loss.max() - loss.min())
    plt.plot(loss_normalized, linestyle='dashed', label=f"{labels[idx]} loss (normalized)")
plt.legend()
plt.xlabel('Epoch')
plt.xlim(0,199)
plt.ylabel('Accuracy / Loss (normalized)')
plt.ylim(top=1.0)
plt.suptitle("Training metrics over 200 epochs")
plt.title("(ConvMixer256/8 vs. ResNet/S-38)")
plt.savefig("../publish/images/final_training.png", dpi=300)
plt.show()
```

```python
plt.figure()
samples = [final_results, resnset_s_results]
labels =['ConvMixer256/8 (k=5, p=2)', 'ResNet/S-38 (n=6)']
for idx, experiment in enumerate(samples):
    plt.plot(data_final[idx].val_acc, label=labels[idx])
    print(f"{labels[idx]} max: {data_final[idx].val_acc.max()}")
plt.legend()
plt.xlabel('Epoch')
plt.xlim(0,199)
plt.ylabel('Accuracy')
plt.ylim(top=1.0)
plt.suptitle("Validation accuracy over 200 epochs")
plt.title("(ConvMixer256/8 vs. ResNet/S-38)")
plt.savefig("../publish/images/final_validation.png", dpi=300)
plt.show()
```
